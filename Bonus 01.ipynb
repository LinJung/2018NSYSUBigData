# Image Convolution

藉由 image convolution 的操作來說明 Convolutional Neural Networks 的 feature extraction 部分的運作原理。

```python
%matplotlib inline
import numpy as np
from scipy import ndimage
from PIL import Image
import matplotlib.pyplot as plt
```

## 讀取圖形

```python
# 讀取影像
img1 = Image.open('lena.jpg')
img2 = Image.open('valve.png')
img3 = Image.open('Bikesgray.jpg')
```

```python
img1
```

```python
img1.size
```

```python
img1.mode
```

```python
# 將彩色影像轉換成灰階影像
img1_gray = img1.convert('L')
```

```python
img1_gray.mode
```

```python
img1_array = np.array(img1_gray)
```

```python
plt.imshow(img1_array, cmap='gray');
```

```python
img1_array.shape
```

#### # 將 img1_array 做一點處理，防止 data overflow...

```python
img1_array.dtype
```

```python
img1_array = img1_array.astype('int64')
```

## 試試看 sobel filter 的效果

```python
sobelx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]], dtype = np.float)
```

```python
sobely = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], dtype = np.float)
```

```python
# Convolve
outx = ndimage.convolve(img1_array, sobelx).astype('float')
```

```python
plt.imshow(outx, cmap='gray');
```

```python
outy = ndimage.convolve(img1_array, sobely).astype('float')
```

```python
plt.imshow(outy, cmap='gray');
```

```python
out = np.sqrt(outx*outx+outy*outy)
```

```python
plt.imshow(out, cmap='gray');
```
